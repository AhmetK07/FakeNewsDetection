```python

#Removing  Hyperlinks
import re
import string
import demoji as dj

ConcatedFake['text'] = ConcatedFake['text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])
ConcatedReel['text'] = ConcatedReel['text'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])


#Removing Punctuations
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

ConcatedReel['text'] = ConcatedReel['text'].apply(lambda x: remove_punct(x))
ConcatedFake['text'] = ConcatedFake['text'].apply(lambda x: remove_punct(x))


#Removing emojis in tweets using Demoji library
ConcatedReel['text'] = ConcatedReel['text'].apply(lambda x: dj.replace(x,""))
ConcatedFake['text'] = ConcatedFake['text'].apply(lambda x: dj.replace(x,""))



#Tokenization
ConcatedReel['text_token'] = ConcatedReel['text'].apply(lambda x: re.split('\W+',x))
ConcatedFake['text_token'] = ConcatedFake['text'].apply(lambda x: re.split('\W+',x))
ConcatedFake["text_token"].values


```

```python
#STOP WORD REMOVING OPERATIONS
with open('turkce-stop-words.txt', 'r', encoding='utf-8') as f:
    stopwords = f.read().split()

ConcatedReel['text_token'] = ConcatedReel['text_token'].apply(lambda tokens: [token for token in tokens if token.lower() not in stopwords and len(token) > 0])
ConcatedFake['text_token'] = ConcatedFake['text_token'].apply(lambda tokens: [token for token in tokens if token.lower() not in stopwords and len(token) > 0])


ConcatedReel['text_token'].values
ConcatedFake['text_token'].values
```

```python
#LEMMATIZER

from typing import List
from jpype import JClass, JString, getDefaultJVMPath, shutdownJVM, startJVM,isJVMStarted


if isJVMStarted():
    # If the JVM is already started, shut it down first
    shutdownJVM()

class TurkishLemmatizer:
    def __init__(self):
        classpath = "zemberek-full.jar"
        # Start the JVM with the classpath
        startJVM(getDefaultJVMPath(), "-Djava.class.path=%s" % classpath)
        # Load the TurkishMorphology class
        self.zemberek = JClass("zemberek.morphology.TurkishMorphology").createWithDefaults()

    def __del__(self):
        # Shutdown the JVM when the object is deleted
        shutdownJVM()

    def lemmatize(self, word: str) -> List[str]:
        # Convert the word to a Java string
        java_word = JString(word)

        # Analyze the word using the Zemberek library
        analysis = self.zemberek.analyze(java_word)

        # Get the lemma for each analysis
        lemmas = []
        for analysis in analysis:
            lemmas.append(analysis.getLemmas()[0])

        return lemmas
    

lemmatizer = TurkishLemmatizer()
```

```python
# Define a function to convert a list of Java strings to Python strings
def java_to_py_string(java_str):
    return str(java_str)

# Define a function to lemmatize a list of tokens
def lemmatize_tokens(tokens):
    lemmatized_tokens = []
    for token in tokens:
        lemmas = lemmatizer.lemmatize(token)
        # Convert the Java strings to Python strings
        lemmas = [java_to_py_string(lemma) for lemma in lemmas]
        if len(lemmas) > 0:
            lemmatized_tokens.append(lemmas[0])
    return lemmatized_tokens

# Apply the lemmatizer to the text_token column in both dataframes
ConcatedReel['text_token_lemmatized'] = ConcatedReel['text_token'].apply(lambda tokens: lemmatize_tokens(tokens))
ConcatedFake['text_token_lemmatized'] = ConcatedFake['text_token'].apply(lambda tokens: lemmatize_tokens(tokens))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Concatenate the text_token_lemmatized lists for each row into a single string
ConcatedReel['text_processed'] = ConcatedReel['text_token_lemmatized'].apply(lambda x: ' '.join(x))
ConcatedFake['text_processed'] = ConcatedFake['text_token_lemmatized'].apply(lambda x: ' '.join(x))

# Define the vectorizer with your desired parameters
vectorizer = TfidfVectorizer(use_idf=True, smooth_idf=True, norm=None, decode_error='replace', analyzer='word', stop_words=stopwords)

# Fit the vectorizer on the concatenated text
vectorizer.fit(ConcatedReel['text_processed'].values.tolist() + ConcatedFake['text_processed'].values.tolist())

# Get the feature names
feature_names = list(vectorizer.vocabulary_.keys())

# Transform the text into TF-IDF vectors
Reel_TFIDF = vectorizer.transform(ConcatedReel['text_processed'].values.tolist())
Fake_TFIDF = vectorizer.transform(ConcatedFake['text_processed'].values.tolist())

# Get the TF-IDF values for each term in the vocabulary
Reel_TFIDF_values = np.sum(Reel_TFIDF, axis=0)
Fake_TFIDF_values = np.sum(Fake_TFIDF, axis=0)

# Convert the values to a numpy array and transpose it to get a row for each term in the vocabulary
Reel_TFIDF_array = np.squeeze(np.asarray(Reel_TFIDF_values))
Fake_TFIDF_array = np.squeeze(np.asarray(Fake_TFIDF_values))

# Create a dataframe with the vocabulary and TF-IDF values for each class
Reel_TFIDF_df = pd.DataFrame({'term': feature_names, 'tfidf': Reel_TFIDF_array})
Fake_TFIDF_df = pd.DataFrame({'term': feature_names, 'tfidf': Fake_TFIDF_array})

# Sort the dataframes by TF-IDF value in descending order
Reel_TFIDF_df = Reel_TFIDF_df.sort_values('tfidf', ascending=False).reset_index(drop=True)
Fake_TFIDF_df = Fake_TFIDF_df.sort_values('tfidf', ascending=False).reset_index(drop=True)
```

```python
from keras.layers import LSTM, Dense, Embedding, Dropout
from keras.models import Sequential
from keras.preprocessing.text import Tokenizer
from keras.callbacks import EarlyStopping
from keras.regularizers import L1L2
from keras.optimizers import Adam
from keras.utils import pad_sequences
from keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split


# Create Word2Vec model
w2v_model = Word2Vec(sentences=ConcatedReel['text_token_lemmatized'].tolist() + ConcatedFake['text_token_lemmatized'].tolist(), vector_size=100, min_count=1, workers=4)

# Tokenize text and convert to sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(ConcatedReel['text_processed'].tolist() + ConcatedFake['text_processed'].tolist())
sequences = tokenizer.texts_to_sequences(ConcatedReel['text_processed'].tolist() + ConcatedFake['text_processed'].tolist())

# Pad sequences to be the same length
padded_sequences = pad_sequences(sequences, maxlen=1000, padding='post')

# Split data back into real and fake
Reel_padded_sequences = padded_sequences[:len(ConcatedReel)]
Fake_padded_sequences = padded_sequences[len(ConcatedReel):]


# Define LSTM model
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=100, input_length=1000, weights=[w2v_model.wv.vectors[:len(tokenizer.word_index) + 1]]))
model.add(LSTM(units=50, dropout=0.3, recurrent_dropout=0.5, kernel_regularizer=L1L2(l1=1e-6, l2=1e-5), recurrent_regularizer=L1L2(l1=1e-6, l2=1e-5), bias_regularizer=L1L2(l1=1e-7, l2=1e-6)))
model.add(Dropout(0.3))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer=Adam(learning_rate=0.005), loss='binary_crossentropy', metrics=['accuracy'])

# Define early stopping and model checkpointing
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)

# Concatenate the real and fake padded sequences and create corresponding labels
all_padded_sequences = np.concatenate([Reel_padded_sequences, Fake_padded_sequences])
all_labels = np.concatenate([np.ones(len(ConcatedReel)), np.zeros(len(ConcatedFake))])

# Split the data into training (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(all_padded_sequences, all_labels, test_size=0.2, random_state=5, stratify=all_labels)

# Train the model on the training set and validate on the test set
model.fit(X_train, y_train, epochs=5, batch_size=64, validation_split=0.2, callbacks=[early_stopping, model_checkpoint])

# Evaluate the model on the test set
test_eval = model.evaluate(X_test, y_test)
print('Test accuracy: {}'.format(test_eval[1]))
```
